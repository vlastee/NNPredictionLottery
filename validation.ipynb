```python
from keras.models import load_model
import numpy as np

model = load_model('PowerBall/weights.600-20.81.hdf5')
x_val = []
one_line_template = [0]*70
x_val.append(list(map(int, one_line_template)))
val_data = [5, 26, 36, 64, 69, 19]
counter_x_val = 0
for x in val_data:
    print("val_data:")
    print(x)
    print('counter_x_val:', counter_x_val)
    print(x)
    x_val[counter_x_val][x-1] = (1)
    print('Added to x_val: ',  x_val[counter_x_val])
        
    print('processed 1 set of numbers -------------------------------------- \n')
    


x_val = np.array(x_val)
print("x_val[]:")
print(x_val)

classes = model.predict(x_val, batch_size=None, verbose=0, steps=None)

print(classes)
for x in classes:
    print(x)
```
